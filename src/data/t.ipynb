```python
def dummy_fun(doc):
    return doc
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(preprocessor= dummy_fun,
tokenizer= dummy_fun,
token_pattern= None)
```

```python
vect.fit(df['lemmas'])
```

```python
vocab = vect.get_feature_names()
vocab_2 = vect.vocabulary_
```

```python
pickle.dump(vocab, open('vectorizer_vocab_1', 'wb'))
pickle.dump(vocab_2, open('vectorizer_vocab_2', 'wb'))
```

```python
vect_1 = TfidfVectorizer(preprocessor=dummy_fun, 
tokenizer = dummy_fun,
token_pattern=None,
vocabulary= pickle.load(open('vectorizer_vocab_1', 'rb')))
vect_1.fit(df['lemmas'])
```

```python
vect_2 = TfidfVectorizer(preprocessor=dummy_fun, 
tokenizer = dummy_fun,
token_pattern=None,
vocabulary= pickle.load(open('vectorizer_vocab_2', 'rb')))
vect_2.fit(df['lemmas'])
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=10)
model.fit(vect_1.transform(df['lemmas']))
```

```python
doc = vect.transform(df['lemmas'][df['name'] == 'Sol Ring'])
_, n = model.kneighbors(doc)
```

```python
for i in n:
    print(df['name'][i])
```
