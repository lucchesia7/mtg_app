```python
import pickle
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def dummy_fun(doc):
    return doc
```

```python
df = pd.read_csv(r'C:\Users\Alex Lucchesi\OneDrive\Documents\GitHub\MTG_app\src\data\oracle_data.csv')
```

```python
vect = TfidfVectorizer(preprocessor= dummy_fun,
                       tokenizer= dummy_fun,
                       token_pattern=None)
vect.fit(df['lemmas'])
doc = vect.transform(df['lemmas'])
nn = NearestNeighbors(n_neighbors=10)
nn.fit(doc)
```

```python
doc = vect.transform(df['lemmas'][df['name'] == 'Blood on the Snow'])
n_index = nn.kneighbors(doc, n_neighbors=11, return_distance=False)
```

```python
vocab = vect.vocabulary_
with open('vectorizer_vocab', 'wb') as f:
        pickle.dump(vocab, f)
```

```python
with open('model', 'wb') as f:
        pickle.dump(nn, f)
```

```python
"string.split(). if split.islower(): if split_string is not [list of stop words], uppercase them; else: if split string is [list of stop words]: lowercase the stopword"



stop_words = ['on', 'the']
cap_stop_words = [w.capitalize() for w in stop_words]
s = ''
card_name = 'Blood On the Snow'
split_name = card_name.split()

for name in split_name:
    if name[0].islower() and name not in stop_words:
        name = name.capitalize()
    elif name[0].isupper() and name in cap_stop_words:
        name = name.lower()
    s += (' ' +name)
    s = s.strip()
print(s)
```
